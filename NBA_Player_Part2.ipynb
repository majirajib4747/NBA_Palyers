```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
# Create engine using the `demographics.sqlite` database file
engine =  sqlalchemy.create_engine('postgresql://postgres:Rajibmaji1#@localhost:5432/nba_db')

# Declare a Base using `automap_base()`
Base = automap_base()

# Use the Base class to reflect the database tables
Base.prepare(autoload_with=engine)

# Print all of the classes mapped to the Base
Base.classes.keys()
#nba_players = Base.classes.nba_players
```

```python
# Assign the demographics class to a variable called `Demographics`
nba_player = Base.classes.nba_player

# Create a session
session = Session(engine)
name = []
dob = []
nba_debut = []
career_length = []

nba_player_dict = { "name" : [],
                    "dob" : [],
                    "nba_debut" : [],
                   "career_length" : []
    
}

# Use the session to query Demographics table and display the first 5 locations
for row in session.query(nba_player).all():
    name.append(row.name)
    dob.append(row.dob)
    nba_debut.append(row.nba_debut)
    career_length.append(row.career_length)
    
nba_player_dict["name"]   = name
nba_player_dict["dob"]   = dob
nba_player_dict["nba_debut"]   = nba_debut
nba_player_dict["career_length"]   = career_length

print(nba_player_dict)
```
